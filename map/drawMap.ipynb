````python
# Import libraries
import pandas as pd
import folium
 
# Load the shape of the zone (US states)
# Find the original file here: https://github.com/python-visualization/folium/tree/master/examples/data
state_geo = 'us-states.json'
 
# read the csv as df and drop data which has minus value in bedcounts
df = pd.read_csv('../MAIN DATA/main_data.csv')
df = df[df['BEDS']>0]


# Load the values of each state
state_data = pd.DataFrame()
state_data['State'] = list(df.groupby('Provider State').mean().index)
state_data['Cost'] = list(df.groupby('Provider State').mean()['Cost'])
state_data['HospitalCounts'] = list(df.groupby('Provider State').count()['Cost'])
state_data['Rating'] = list(df.groupby('Provider State').mean()['Hospital overall rating'])
state_data['BedCounts'] = list(df.groupby('Provider State').mean()['BEDS'])
# state_data
threshold_scale = split_six(state_data['Cost'])
colors = ['#1a9850', '#91cf60', '#fc8d59', '#d9ef8b', '#ffffbf','#fee08b']

# Initialize the map:
m = folium.Map(location=[37, -102], zoom_start=5)
# step = folium.StepColormap(colors, vmin=921, vmax=2022, index=[921, 1105, 1288, 1472, 1655,1838,2022])
# step = folium.StepColormap(colors)

# Add the color for the chloropleth:
m.choropleth(
 geo_data=state_geo,
 name='choropleth',
 data=state_data,
 columns=['State', 'Cost'],
 key_on='feature.id',
 fill_color='RdYlGn',
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='Cost ($)'
)
folium.LayerControl().add_to(m)

m
````
