```python
import pandas as pd
import numpy as np
```

```python
hospital = "Inpatient_Prospective_Payment_System.csv"
data = pd.read_csv(hospital)


top_5 = data['DRG Definition'].value_counts()
top_5 = top_5.head(5)
```

```python
top_5 = top_5.reset_index()
```

```python

top_5 = top_5.rename(columns={"index":"DRG Definition", "DRG Definition":"count"})
top_5
```

```python
newdf = pd.merge(top_5, data, how=  "inner", on = "DRG Definition")
newdf
```
