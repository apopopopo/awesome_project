```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("main_cleaned.csv")
df
del df[df.columns[0]]
df
```

```python
df.columns
```

```python
rating_gov_df = df.loc[df["Hospital Ownership"]=="Government"]

rating_gov_df
```

```python
rating_business_df=df.loc[df["Hospital Ownership"]=="Individual Business"]
rating_business_df
```

```python
rating_voluntary_df=df.loc[df["Hospital Ownership"]=="Voluntary(non-profit)"]
rating_voluntary_df.columns
```

```python
rating_voluntary=rating_voluntary_df["Hospital overall rating"].mean()
rating_voluntary
```

```python
rating_business=rating_business_df["Hospital overall rating"].mean()
```

```python
rating_gov=rating_gov_df["Hospital overall rating"].mean()
```

```python
fig, ax = plt.subplots()
ratings=[rating_gov,rating_business,rating_voluntary]
ind = np.arange(len(ratings))  
width=0.75

gov =ax.bar(ind[0], ratings[0], width, color='red')
business=ax.bar(ind[1], ratings[1], width, color='blue')
voluntary=ax.bar(ind[2], ratings[2], width, color='yellow')

ax.set_ylabel('Ratings')
ax.set_title('Ratings VS Ownership')
ax.set_xticks(ind)
ax.set_xticklabels(('Government', 'Business', 'Voluntary'))
ax.set_autoscaley_on(True)
ax.grid(True)
for i, v in enumerate(ratings):
    plt.text(i-.12,v+.01, f'{round(v, 2)}', color='black', fontweight='bold', fontsize=10)
```
