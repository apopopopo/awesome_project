```python
import plotly.plotly as py
import plotly.tools as tls
import pandas as pd
import numpy as np
from pprint import pprint
import requests
import json
from config import gkey
```

```python
df = pd.read_csv("../top_5_services.csv")
print(df.columns)
df.head(3)
```

```python
df_less = df[['Provider Id', 'Provider Name', 'Provider Street Address', 'Provider City', 'Provider State', 'Hospital Referral Region Description']]
df_less
```

```python

grouped = df.groupby('Provider Id').mean()
new = pd.DataFrame(grouped).reset_index(inplace=False)

merged_df = df_less.merge(new, on='Provider Id', how='inner', copy=False)
print(len(merged_df))
merged_df
```

```python
df = merged_df.drop_duplicates(keep = 'first').reset_index(inplace=False)
df = df.drop(['index', 'Unnamed: 0'], axis=1)
df.head()
```

```python
#join address and city and state
address = []

def get_address(index):
    target = df['Provider Street Address'][index] + ", " + df['Provider City'][index] + ", " + df['Provider State'][index]
    return target
for i in range(len(df)):
    address.append(get_address(i))
```

```python
#return json response
lat = []
lng = []
counter = 0
for target_address in address:
    params = {"address": target_address, "key": gkey}
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    response = requests.get(base_url, params=params).json()
    #traverse response to find lon/lat
    try:
        lat.append(response['results'][0]['geometry']['location']['lat'])
        lng.append(response['results'][0]['geometry']['location']['lng'])
    except Exception:
        lat.append(0)
        lng.append(0)
    counter = counter + 1
    print(f'{counter} of {len(df)} completed')
#return lon/lat 
print(lat,lng)
```

```python
#add to df
df['Latitude'] = lat
df['Longitude'] = lng
df.to_csv('USA costs with latlon.csv')
df.head(2)
```

```python
#input into plotly by scattergeo
data = [dict(type="scattergeo", locationmode="USA-states", text = df['Provider Name'] + df['Provider Street Address'], lon = df['Longitude'], lat = df['Latitude'], 
             marker = dict(size=6, autocolorscale=False, colorscale='custom-colorscale',
            color = df['Average Medicare Payments'], colorbar = dict(title = 'Avg Cost ($)')))]
```

```python
layout = dict(title="Avg Cost by Hospitals in US",
              geo = dict(scope='usa',projection = dict(type='albers usa'), 
                         showland=True, landcolor="rgb(250,250,250)"))
```

```python
fig = dict(data=data, layout=layout)
py.iplot(fig)
# py.image.save_as(fig,filename='USA_plot.png')
```
